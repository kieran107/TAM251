```python
from sympy import *
import numpy as np

k1 = 271 # N/mm
k2 = 195 # N/mm
L = 200 # mm
theta = 27 # degrees
P = 17 # N

t = np.radians(theta)
d1, d2 = symbols('d1, d2')
eqn1 = P*1 - k2*d2*2 - k1*d1*cos(t)*2
eqn2 = d2*cos(t) - d1
solve((eqn1, eqn2), (d1, d2))
```

# HW4.2

```python
from sympy import *

h = 110  # mm
b = 101  # mm
d = 28  # mm
E1 = 157  # MPa
E2 = 213  # MPa
W = 95  # N

t1 = atan(h/b)
t2 = atan(h/(b*2))
b1 = pi/2 - t1
b2 = pi/2 - t2

d1od2 = cos(b1) / (2*cos(b2))

L1 = h / sin(t1)
L2 = h / sin(t2)
A = pi/4*d**2
f1 = L1 / (A*E1)
f2 = L2 / (A*E2)

F1oF2 = d1od2 * (E1*L2) / (E2*L1)

F1, F2 = symbols('F1, F2')
eqn1 = F1/F2 - F1oF2
eqn2 = F1*sin(t1) + F2*sin(t2)*2 - W/2 - 2*W/2
ans = solve((eqn1, eqn2), (F1, F2))

s1 = float(ans[F1]/A)
s2 = float(ans[F2]/A)

d1od2, f1.evalf(), f2.evalf(), F1oF2, s1, s2
```

# HW4.3

```python
from sympy import *

E1 = 69 # GPa
E2 = 201 # GPa
d = 27 # mm
D = 41 # mm
L = 800 # mm
P = 86 # kN

A1 = pi/4*d**2
A2 = pi/4*D**2 - A1

F1, F2 = symbols('F1, F2')
eqn1 = F1 + F2 - P
eqn2 = F1*L / (A1*E1) - F2*L / (A2*E2)
ans = solve((eqn1, eqn2), (F1, F2))

s = -float(ans[F2]/A2)
s*1e3
```

# HW4.5

```python
from sympy import *

E1 = 74 # GPa
E2 = 199 # GPa
A = 1763 # mm^2
L = 728 # mm
deltaM = 0.7 # mm

d1, d2 = symbols('d1, d2')
L1 = L
L2 = deltaM + L

eqn1 = d1 + d2 - deltaM
eqn2 = d1*E1*A/L1 - d2*E2*A/L2
ans = solve((eqn1, eqn2), (d1, d2))

ep = -ans[d1] / L
ep*1e3
```

```python
from sympy import *

deltaM = 3 # mm
k1 = 41000 # N/mm
k2 = 12000 # N/mm
L = 2200 # mm

d = symbols('d')
eqn = k1*d - (deltaM - d)*k2
ans = solve(eqn, d)

d1 = float(ans[0])
d2 = deltaM - d1

ep1 = d1 / L
ep2 = -d2 / (L + deltaM)

ep1*1e3, ep2*1e3
```

```python
from sympy import *

dBrass = 30 # mm
dSteel = 12 # mm
Ls = 791 # mm
Lb = 719 # mm
Es = 165 # GPa
Eb = 86 # GPa
P = 40 # kN

A1 = pi/4*dBrass**2 * 2
A2 = pi/4*dSteel**2

N1, N2 = symbols('N1, N2')
eqn1 = N1 + N2 - P
eqn2 = N1*Lb / (A1*Eb) - N2*Ls / (A2*Es)
ans = solve((eqn1, eqn2), (N1, N2))

F1 = -ans[N1]
s1 = F1 / A1 *1e3

F2 = -ans[N2]
s2 = F2 / A2 *1e3

s1.evalf(), s2.evalf()
```
