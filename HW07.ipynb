```python
from sympy import *

h = 141 # mm
b = 243 # mm
t1 = 24 # mm
t2 = 21 # mm
Mz = 1342 # N.m

H = 2*t2+h
B = 2*t1+b

a = h*b
A = H*B

ybar = (A*H/2 - a*(t2+h/2)) / (A - a)

D = H/2 - ybar
d = D

Ii = 1/12*b*h**3 + a*d**2
Io = 1/12*B*H**3 + A*D**2
I = Io - Ii

c = H - ybar
sigma = Mz*c / I

ybar, I/10**6, sigma*10**3
```

```python
from sympy import *

R = 25 # mm
Mz = 1634 # N.m

H = R
B = pi*R
A = H*B
a = 3*pi*R**2

ybar = (A * 2.5*R)/(A + a)

D = 2.5*R - ybar
d = ybar

Ir = 1/12*B*H**3 + A*D**2 
Ii = 61*pi*R**4/12 + a*d**2 

I = Ir + Ii

c = 3*R - ybar

ybar, (I/10**6).evalf(), abs(Mz*c/I).evalf()*10**3
```
