```python
from numpy import *

P = 3 # kN
a = 562 # mm
b = 585 # mm
L = 1089 # mm
E = 183 # GPa
Ac = 891 # mm^2
theta = 42 # degrees

theta = radians(theta)
By = P*L / a
F = By / sin(theta)
s = F / Ac 
ep  = s / E

DB = b*(1 + ep)
dB = sqrt(DB**2 - (b*cos(theta))**2) - b*sin(theta)

dB/a*L
```
